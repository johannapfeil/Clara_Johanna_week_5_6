```python
# Basic inspecting of the dataframe
def inspect_dataframe(merged_df):
    """
    Function to perform basic inspection on a DataFrame: 
    shape, column names, data types, and missing values.
    
    """

    print('Check the shape (rows, columns):')
    print(merged_df.shape)

    print('\nColumn names:')
    print(merged_df.columns)


    print('\nData types:')
    print(merged_df.dtypes)


    print('\nMissing values:')
    print(merged_df.isnull().sum())

# clean column names
def clean_column_names(merged_df):
    """
    Function to clean the column names of a DataFrame:
    - Convert to lowercase
    - Replace spaces with underscores
    - Remove or replace special characters with underscores

    """

    def clean_name(name):
        name = name.lower()
        name = name.replace(" ", "_")
        name = re.sub(r'[^a-z0-9_]', '_', name)
        return name
    
    merged_df.columns = [clean_name(col) for col in merged_df.columns]    
    return merged_df

# check unique and empty values
def check_unique_and_empty(merged_df):
    """
    Function to print the unique and empty values for each column in a DataFrame.
    
    """
    result = []
    
    for column in merged_df.columns:
        unique_values = merged_df[column].nunique()
        empty_values = merged_df[column].isna().sum()
        
        empty_rows = merged_df[column][merged_df[column].isna()].index.tolist()
        
        result.append({
            'Column': column,
            'Unique Values': unique_values,
            'Empty Values': empty_values,
            'Empty Row Indices': empty_rows
        })
    
    merged_df = pd.DataFrame(result)
    
    merged_df.set_index('Column', inplace=True)
    
print(inspect_dataframe(merged_df)) 
print(clean_column_names(merged_df))
print(check_unique_and_empty(merged_df))
display(merged_df)
```

```python
# Identify and keep only the last occurrence of consecutive same steps
merged_df['prev_process_step'] = merged_df.groupby('visit_id')['process_step'].shift(1)

# Filter to keep only rows where the current process_step is different from the previous one,
# or it's the first occurrence in the group
merged_df = merged_df[merged_df['process_step'] != merged_df['prev_process_step']]

# Drop the helper column used for comparison
merged_df = merged_df.drop(columns=['prev_process_step'])

# Reset index if needed
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
# Identifying and dropping the visit_ids that don't have 'start' in the process_step column
visits_with_start = merged_df[merged_df['process_step'] == 'start']['visit_id'].unique()
filtered_df_with_start = merged_df[merged_df['visit_id'].isin(visits_with_start)]
display(filtered_df_with_start)

# Identifying and dropping the client_ids that don't have 'start' in the process_step column
clients_with_start = filtered_df_with_start[filtered_df_with_start['process_step'] == 'start']['client_id'].unique()
filtered_df = filtered_df_with_start[filtered_df_with_start['client_id'].isin(clients_with_start)]
display(filtered_df)
```

```python
# Step 1: Identify 'start' steps
start_steps = filtered_df[filtered_df['process_step'] == 'start']

# Step 2: Find 'start' steps followed by 'step_1'
# Create a column to check the next step within each visit_id
filtered_df['next_step'] = filtered_df.groupby('visit_id')['process_step'].shift(-1)

# Keep 'start' steps directly followed by 'step_1'
start_followed_by_step_1 = filtered_df[(filtered_df['process_step'] == 'start') & 
                                       (filtered_df['next_step'] == 'step_1')]

# Step 3: Handle remaining 'start' steps not followed by 'step_1'
# Get the visit_ids where 'start' isn't followed by 'step_1'
remaining_start_steps = start_steps[~start_steps['visit_id'].isin(start_followed_by_step_1['visit_id'])]

# For visit_ids with multiple 'start' steps, keep only the latest one
latest_start_per_visit = remaining_start_steps.loc[remaining_start_steps.groupby('visit_id')['date_time'].idxmax()]

# Combine the filtered 'start' steps
final_start_steps = pd.concat([start_followed_by_step_1, latest_start_per_visit])

# Step 4: Merge the filtered 'start' steps back with the original DataFrame
# Keep all rows from the original DataFrame, but only include the filtered 'start' steps
filtered_df = pd.concat([filtered_df[filtered_df['process_step'] != 'start'], final_start_steps]).sort_values(by=['visit_id', 'date_time'])

# Clean up the temporary 'next_step' column
filtered_df = filtered_df.drop(columns=['next_step'])
filtered_df
```

```python
# Step 1: Create a column to track the next step for each row within the same visit_id
filtered_df['next_step'] = filtered_df.groupby('visit_id')['process_step'].shift(-1)

# Step 2: Identify 'confirm' steps directly followed by another 'confirm'
filtered_df['is_skippable_confirm'] = (filtered_df['process_step'] == 'confirm') & (filtered_df['next_step'] == 'confirm')

# Step 3: Mark the last 'confirm' step in cases where there are consecutive 'confirm' steps
filtered_df['is_last_confirm'] = (filtered_df['process_step'] == 'confirm') & \
                                 (filtered_df['next_step'] != 'confirm')

# Step 4: Keep all rows, but only the last 'confirm' step in consecutive cases
filtered_df = filtered_df[(filtered_df['process_step'] != 'confirm') | 
                          (~filtered_df['is_skippable_confirm']) | 
                          (filtered_df['is_last_confirm'])]

# Step 5: Drop temporary columns used for filtering
filtered_df = filtered_df.drop(columns=['next_step', 'is_skippable_confirm', 'is_last_confirm'])

filtered_df
```

```python
# Define the valid steps order
valid_steps_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Map the steps to their order for easier comparison
step_to_order = {step: idx for idx, step in enumerate(valid_steps_order)}

def is_valid_sequence(group):
    # Map steps to their order
    step_orders = group['process_step'].map(step_to_order).tolist()
    
    # Check if the sequence matches forward or backward order without skipping
    for i in range(len(step_orders) - 1):
        if abs(step_orders[i] - step_orders[i + 1]) != 1:
            return False  # Skipped steps found
    return True

# Filter the DataFrame for valid consecutive steps (forward or backward)
filtered_df = (
    filtered_df.groupby('visit_id', group_keys=False)
    .filter(is_valid_sequence))

filtered_df
```

### For our project, we need to know how much time a client spends on each step and if they ever reached the last step.

```python
# Define the valid order of process steps
valid_steps_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Initialize a list to store the results
time_spent = []

# Function to map steps to their order
def step_order(step):
    return valid_steps_order.index(step)

# Iterate over each unique combination of client, visitor, and visit
for (client_id, visitor_id, visit_id), group in filtered_df.groupby(['client_id', 'visitor_id', 'visit_id']):
    prev_time = None
    prev_step = None
    last_valid_step_order = -1  # Start with an invalid step order to force a check for the first step
    
    for _, row in group.iterrows():
        current_time = pd.to_datetime(row['date_time'])
        current_step = row['process_step']
        
        # Check if this is the first step
        if prev_step is None:
            # Only process if the current step is 'start'
            if current_step == 'start':
                time_spent.append({
                    'client_id': client_id,
                    'visitor_id': visitor_id,
                    'visit_id': visit_id,
                    'from_step': None,
                    'to_step': current_step,
                    'time_spent': None,
                    'is_error': False  # No error for the first 'start' step
                })
                prev_time = current_time
                prev_step = current_step
                last_valid_step_order = step_order(current_step)
            continue  # Skip the rest for invalid first steps

        # For subsequent steps, check the order and calculate time
        current_step_order = step_order(current_step)
        if current_step_order > last_valid_step_order + 1:
            # If there's a step order skip, drop this row
            continue
        elif current_step_order < last_valid_step_order:
            # Backward step is considered an error
            time_diff = current_time - prev_time
            time_spent.append({
                'client_id': client_id,
                'visitor_id': visitor_id,
                'visit_id': visit_id,
                'from_step': prev_step,
                'to_step': current_step,
                'time_spent': time_diff,
                'is_error': True
            })
        else:
            # Valid transition
            time_diff = current_time - prev_time
            time_spent.append({
                'client_id': client_id,
                'visitor_id': visitor_id,
                'visit_id': visit_id,
                'from_step': prev_step,
                'to_step': current_step,
                'time_spent': time_diff,
                'is_error': False
            })

        # Update previous step info
        prev_time = current_time
        prev_step = current_step
        last_valid_step_order = current_step_order

# Create a DataFrame with the results
time_spent_df = pd.DataFrame(time_spent)

# Drop rows with None values in 'from_step' and 'to_step'
time_spent_df = time_spent_df.dropna(subset=['from_step', 'to_step'])

# Display the resulting DataFrame
time_spent_df
```

```python
# Add a new column to check if 'from_step' and 'to_step' are the same
time_spent_df['is_same_step'] = time_spent_df['from_step'] == time_spent_df['to_step']

# Filter rows where the two columns are the same
same_step_rows = time_spent_df[time_spent_df['is_same_step']]
display(same_step_rows)

# Filter rows where 'from_step' and 'to_step' are the same
same_step_rows = time_spent_df[time_spent_df['from_step'] == time_spent_df['to_step']]

# Drop these rows from the main DataFrame
time_spent_df = time_spent_df.drop(same_step_rows.index)
time_spent_df = time_spent_df.drop(columns=['is_same_step'])

# Ensure 'to_step' is retained in the filtered DataFrame
time_spent_df = time_spent_df.reset_index(drop=True)
display(time_spent_df)
```

```python
errors_occured = time_spent_df['is_error'].value_counts()
errors_occured
```

```python
# Filter transitions to include only those that follow the valid steps order
valid_transitions = []

# Iterate over the time_spent_df and filter valid transitions
for _, row in time_spent_df.iterrows():
    from_step = row['from_step']
    to_step = row['to_step']
    
    # Check if the transition follows the valid steps order (i.e., from a step to the next one in order)
    if from_step is not None and to_step is not None:
        if valid_steps_order.index(to_step) == valid_steps_order.index(from_step) + 1:
            valid_transitions.append(row)

# Create a DataFrame with only valid transitions
valid_transitions_df = pd.DataFrame(valid_transitions)

# Calculate the average time spent on each transition
avg_time_spent = valid_transitions_df.groupby(['from_step', 'to_step'])['time_spent'].mean().round(0).reset_index()

# Rename the columns for clarity
avg_time_spent = avg_time_spent.rename(columns={'time_spent': 'avg_time_spent'})

# Display the resulting DataFrame
avg_time_spent
```

Some clients didn't ever complete the process. Therefore, we need to ensure they are identified and see how many clients reached which step.

```python
# Identify clients who didn't complete the whole process (didn't reach "confirm")
completed_clients = time_spent_df[time_spent_df['to_step'] == 'confirm']['visit_id'].unique()

# Identify clients who didn't reach "confirm"
incomplete_clients = time_spent_df[~time_spent_df['visit_id'].isin(completed_clients)]['visit_id'].unique()

# Show which clients are incomplete
print(f"Clients who didn't complete the process: {incomplete_clients}")

# Check the last step they reached
last_steps = time_spent_df.groupby('visit_id')['to_step'].last()

# Show the last step for each client (whether complete or incomplete)
print(f"Last step for each client: \n{last_steps}")
```

```python
# Get the total number of unique visits
total_visits = filtered_df['visit_id'].nunique()

# Identify the unique steps, excluding 'start' because it didn't make sense to keep it
steps = ['start','step_1', 'step_2', 'step_3', 'confirm'] 

# Initialize a dictionary to store the percentage of clients reaching each step
step_percentages = {}

# Iterate through each step and calculate the percentage
for step in steps:
    visits_reached_step = filtered_df[filtered_df['process_step'] == step]['visit_id'].nunique()
    # Calculate the percentage of visits who reached this step
    step_percentages[step] = (visits_reached_step / total_visits) * 100

# Print the results
for step, percentage in step_percentages.items():
    print(f"Percentage of visits who reached {step}: {percentage:.2f}%")
```
