## Read 1

```python
completion_rate_step_test = pd.read_csv('../data/intermediate_steps/completion_rate_step_test.csv')
completion_rate_step_test.drop('Unnamed: 0', axis = 1, inplace = True)
completion_rate_step_test
```

```python
completion_rate_step_control = pd.read_csv('../data/intermediate_steps/completion_rate_step_control.csv')
completion_rate_step_control.drop('Unnamed: 0', axis = 1, inplace = True)
completion_rate_step_control
```

## Read 2

```python
average_time_spent_test = pd.read_csv('../data/intermediate_steps/average_time_spent_test.csv')
average_time_spent_test.drop('from_step', axis = 1, inplace = True)
average_time_spent_test.drop('Unnamed: 0', axis = 1, inplace = True)
average_time_spent_test
```

```python
average_time_spent_control = pd.read_csv('../data/intermediate_steps/average_time_spent_control.csv')
average_time_spent_control.drop('from_step', axis = 1, inplace = True)
average_time_spent_control.drop('Unnamed: 0', axis = 1, inplace = True)
average_time_spent_control
```

## Read 3 & 4

```python
errors_test = pd.read_csv('../data/intermediate_steps/errors_test.csv')
errors_test.drop('Unnamed: 0', axis = 1, inplace = True)
errors_test
```

```python
errors_control = pd.read_csv('../data/intermediate_steps/errors_control.csv')
errors_control.drop('Unnamed: 0', axis = 1, inplace = True)
errors_control
```

# Make a big data frame

```python
a_test = pd.merge(completion_rate_step_test, average_time_spent_test, on = 'to_step')
b_test = pd.concat([a_test, errors_test])
b_test['Variation'] = 'Test'
b_test.to_csv('../data/clean/results_test.csv')
```

```python
a_control = pd.merge(completion_rate_step_control, average_time_spent_control, on = 'to_step')
b_control = pd.concat([a_control, errors_control])
b_control['Variation'] = 'Control'
b_control.to_csv('../data/clean/results_control.csv')
```

```python
df = pd.concat([b_test, b_control], axis = 0)
df.to_csv('../data/clean/results.csv')
```
